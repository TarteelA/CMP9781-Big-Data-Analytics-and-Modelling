```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Workshop 7").getOrCreate()
```

```python
df = spark.read.csv('walmart_stock.csv', header=True, inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
df.describe('High', 'Low', 'Volume').show()
```

```python
df2 = df.filter(df['Close']<60)
df2.show()
```

```python
df2 = df.filter(df['High']>80).count()/df.count() * 100
df2
```

```python
df.withColumn('HV Ratio', df['High']/df['Volume']).show()
```
