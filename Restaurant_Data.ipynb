```python
#Read Data
import pandas as pd
Data = pd.read_csv('restaurant_data.csv')
Data
```

```python
#Remove Data Records With Empty Values
Data = Data.dropna()
Data
```

```python
#Filtering Out Data with Rating No Less Than 5
Data1 = Data[Data.Rating >= 5]
Data1
```

```python
#Statistical Analysis
DF = pd.DataFrame(Data)
DF.describe()
```

```python
#Correlation Calculation
Correlation = Data['Revenue'].corr(Data['Marketing Budget'])
Correlation
```

```python
#One-Hot Encoding for Name Categorical Variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Data['Name_encoded'] = encoder.fit_transform(Data['Name'])
Data
```

```python
#One-Hot Encoding for Location Categorical Variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Data['Location_encoded'] = encoder.fit_transform(Data['Location'])
Data
```

```python
#One-Hot Encoding for Cuisine Categorical Variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Data['Cuisine_encoded'] = encoder.fit_transform(Data['Cuisine'])
Data
```

```python
#One-Hot Encoding for Parking Availability Categorical Variable
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Data['Parking Availability_encoded'] = encoder.fit_transform(Data['Parking Availability'])
Data
```

```python
#Data Normalisation
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

DF = pd.DataFrame(Data)
cols = ['Rating', 'Seating Capacity', 'Average Meal Price', 'Marketing Budget', 'Social Media Followers', 
        'Chef Experience Years', 'Number of Reviews', 'Avg Review Length', 'Ambience Score', 'Service Quality Score', 
        'Weekend Reservations', 'Weekday Reservations', 'Revenue']
DF = DF.loc[:, cols]

Scaler = MinMaxScaler()
DF_Normalised = pd.DataFrame(Scaler.fit_transform(DF), columns = DF.columns)
DF_Normalised
```

```python
#Preparing Training/Testing Data
Revenue = DF['Revenue'].values.reshape(-1, 1)
Marketing_Budget = DF['Marketing Budget'].values.reshape(-1, 1)

from sklearn.model_selection import train_test_split
Marketing_Budget_Train, Marketing_Budget_Test, Revenue_Train, Revenue_Test = train_test_split(Marketing_Budget, Revenue, test_size = 0.3)
```

```python
Revenue_Train
```

```python
Revenue_Test
```

```python
#Model Training and Testing
from sklearn.linear_model import LinearRegression
Regressor = LinearRegression()
Regressor.fit(Marketing_Budget_Train, Revenue_Train)
Revenue_Pred = Regressor.predict(Marketing_Budget_Test)
```

```python
Revenue_Pred
```

```python
Revenue_Test
```

```python
#Model Evaluation
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
MSE = mean_squared_error(Revenue_Test, Revenue_Pred)
MAE = mean_absolute_error(Revenue_Test, Revenue_Pred)
RMSE = np.sqrt(MSE)
```
